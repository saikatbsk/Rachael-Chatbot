## Explore the Ubuntu Dialog Corpus Dataset

The Ubuntu Dialog Corpus (UDC) ([paper](https://arxiv.org/abs/1506.08909), [github](https://github.com/rkadlec/ubuntu-ranking-dataset-creator)) is one of the largest public dialog datasets available. Itâ€™s based on chat logs from the Ubuntu channels on a public IRC network.

```python
""" Importing our dependencies. """
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import random

plt.style.use('ggplot')
%matplotlib inline
```

```python
""" Loading a subset of the dataset. """
nb_records = 1000000 #Number of records in file
nb_samples = 100000  #Desired sample size
skip = sorted(random.sample(range(1, nb_records+1), nb_records-nb_samples))

train_db = pd.read_csv('../../Datasets/UbuntuDialogCorpus/data/train.csv', skiprows=skip)
train_db.Label = train_db.Label.astype('category')

test_db = pd.read_csv('../../Datasets/UbuntuDialogCorpus/data/test.csv', skiprows=skip)
validation_db = pd.read_csv('../../Datasets/UbuntuDialogCorpus/data/valid.csv', skiprows=skip)
```

```python
train_db.describe()
```

```python
train_db.Label.hist()
plt.title('Training Label Distribution')
plt.show()
```

```python
pd.options.display.max_colwidth = 500
train_db.head()
```

```python
fig = plt.figure()
g = gridspec.GridSpec(1, 2, right=1.5, wspace=0.5)

train_db_context_len = train_db.Context.str.split(' ').apply(len)
fig.add_subplot(g[0])
plt.title("Training Context Length Statistics")
train_db_context_len.hist()

train_db_utterance_len = train_db.Utterance.str.split(' ').apply(len)
fig.add_subplot(g[1])
plt.title("Training Utterance Length Statistics")
train_db_utterance_len.hist()

plt.show()
```
